A pesquisa InstaVarejo surgiu em virtude do crescimento expressivo e simultâneo do uso da rede social Instagram e do faturamento das grandes redes varejistas. Ambos os fenômenos de consumo tomaram maior forma no mês 03 de 2020 e nos meses seguintes.
