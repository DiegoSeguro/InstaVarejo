#### Importação das bibliotecas e dos arquivos de dados

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

df1_institucional = pd.read_csv ('Institucional.csv' , sep =',') 
df2_produto_uso_generico = pd.read_csv ('Produto receita genérica.csv' , sep =',')
df3_produto_uso_saude = pd.read_csv ('Produto receita benefícios saúde.csv' , sep =',')
df4_promocao = pd.read_csv ('Promoção Engajamento.csv' , sep =',')
df5_responsabilidade_socioambiental = pd.read_csv ('Responsabilidade sócioambiental.csv' , sep =',')
df6_datacomemorativa_sazonal = pd.read_csv ('Data comemorativa sazona.csv' , sep =',')
df7_autoajuda = pd.read_csv ('Autoajuda.csv' , sep =',')
df8_influenciadores = pd.read_csv ('Influenciadores.csv' , sep =',')
df9_higiene_pessoal_cosmeticos = pd.read_csv ('Higiene Pessoal e Cosméticos.csv' , sep =',')
df10_marcas_especificas = pd.read_csv ('Divulgação marcas específicas.csv' , sep =',')
df11_unificado = pd.read_csv ('Base _unificada_todas_categorias.csv' , sep =',')
```

#### Alteração do nome das colunas

```python
df1_institucional.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df2_produto_uso_generico.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df3_produto_uso_saude.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df4_promocao.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df5_responsabilidade_socioambiental.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df6_datacomemorativa_sazonal.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df7_autoajuda.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df8_influenciadores.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df9_higiene_pessoal_cosmeticos.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df10_marcas_especificas.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'comentários']
df11_unificado.columns = ['nome da marca', 'posts', 'data de criação' , 'engajamento' , 'número de likes' , 'plataforma', 'categoria']
```

```python

# Transformação das colunas 'Engagement' e 'Number of Like' de Object para Numeric

df1_institucional['engajamento'] = pd.to_numeric(df1_institucional['engajamento'], errors ='coerce')
df1_institucional['número de likes'] = pd.to_numeric(df1_institucional['número de likes'], errors ='coerce')
df2_produto_uso_generico['engajamento'] = pd.to_numeric(df2_produto_uso_generico['engajamento'], errors ='coerce')
df2_produto_uso_generico['número de likes'] = pd.to_numeric(df2_produto_uso_generico['número de likes'], errors ='coerce')
df3_produto_uso_saude['engajamento'] = pd.to_numeric(df3_produto_uso_saude['engajamento'], errors ='coerce')
df3_produto_uso_saude['número de likes'] = pd.to_numeric(df3_produto_uso_saude['número de likes'], errors ='coerce')
df4_promocao['engajamento'] = pd.to_numeric(df4_promocao['engajamento'], errors ='coerce')
df4_promocao['número de likes'] = pd.to_numeric(df4_promocao['número de likes'], errors ='coerce')
df5_responsabilidade_socioambiental['engajamento'] = pd.to_numeric(df5_responsabilidade_socioambiental['engajamento'], errors ='coerce')
df5_responsabilidade_socioambiental['número de likes'] = pd.to_numeric(df5_responsabilidade_socioambiental['número de likes'], errors ='coerce')
df6_datacomemorativa_sazonal['engajamento'] = pd.to_numeric(df6_datacomemorativa_sazonal['engajamento'], errors ='coerce')
df6_datacomemorativa_sazonal['número de likes'] = pd.to_numeric(df6_datacomemorativa_sazonal['número de likes'], errors ='coerce')
df7_autoajuda['engajamento'] = pd.to_numeric(df7_autoajuda['engajamento'], errors ='coerce')
df7_autoajuda['número de likes'] = pd.to_numeric(df7_autoajuda['número de likes'], errors ='coerce')
df8_influenciadores['engajamento'] = pd.to_numeric(df8_influenciadores['engajamento'], errors ='coerce')
df8_influenciadores['número de likes'] = pd.to_numeric(df8_influenciadores['número de likes'], errors ='coerce')
df9_higiene_pessoal_cosmeticos['engajamento'] = pd.to_numeric(df9_higiene_pessoal_cosmeticos['engajamento'], errors ='coerce')
df9_higiene_pessoal_cosmeticos['número de likes'] = pd.to_numeric(df9_higiene_pessoal_cosmeticos['número de likes'], errors ='coerce')
df10_marcas_especificas['engajamento'] = pd.to_numeric(df10_marcas_especificas['engajamento'], errors ='coerce')
df10_marcas_especificas['número de likes'] = pd.to_numeric(df10_marcas_especificas['número de likes'], errors ='coerce')
df11_unificado['engajamento'] = pd.to_numeric(df11_unificado['engajamento'], errors ='coerce')
df11_unificado['número de likes'] = pd.to_numeric(df11_unificado['número de likes'], errors ='coerce')
```
